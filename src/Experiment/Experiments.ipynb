# 1. Data collection

```python
import pandas as pd
```

```python
data = pd.read_csv("rent_apartments.csv")
```

```python
data
```

# 2. Data preparation

```python
data.dtypes
```

```python
data_encoded = pd.get_dummies(data, 
                              columns = ['balcony',
                                         'parking', 
                                         'furnished', 
                                         'garage', 
                                         'storage'], 
                              drop_first=True)
```

```python
data_encoded
```

```python
data_encoded.garden.unique()
```

```python
import re
```

```python
data_encoded.garden[3]
```

for i in range(len(data_encoded)):
    if data_encoded.garden[i]=='Not present':
        data_encoded.garden[i]=0
    else: data_encoded.garden[i] = int(re.findall(r'\d+', data_encoded.garden[i])[0])

```python
data_encoded.garden.unique()
```

# 3. Model Building

## 3.1. Defining X and y

```python
X = data_encoded[['area', 
                  'constraction_year', 
                  'bedrooms', 
                  'garden', 
                  'balcony_yes', 
                  'parking_yes', 
                  'furnished_yes', 
                  'garage_yes', 
                  'storage_yes']]
```

```python
y = data_encoded.rent
```

```python
y
```

```python
X
```

## 3.2. Split the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## 3.3. Model Building

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(X_train, y_train)
```

## 3.4. Evaluate Model

```python
rf.score(X_test, y_test)
```

## 3.5. Predicting

```python
X
```

```python
rf.predict([[85, 2015, 2, 20, 1, 1, 0, 0, 1]])
```

## 3.6. Tuning Hyperparameters

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_space = {'n_estimators': [100, 200, 300], 'max_depth': [3, 6, 9, 12]}
```

```python
grid = GridSearchCV(RandomForestRegressor(), param_grid=grid_space, cv=5, scoring = 'r2')
```

```python
model_grid = grid.fit(X_train, y_train)
```

```python
print(f'Best hyperparameters are {model_grid.best_params_}, score = {model_grid.best_score_}')
```

```python
model_grid.best_estimator_.predict([[85, 2015, 2, 20, 1, 1, 0, 0, 1]])
```
